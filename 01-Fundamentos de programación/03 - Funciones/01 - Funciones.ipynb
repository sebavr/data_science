```python
# evaluar una solicitud de credito
# 1- mayor de 25
# 2- no tener deuda superior a la mitad del monto solicitad
# 3- no exceder 3 veces el ingreso
# 4- y si soy mayor de 65, no tener deuda
def evaluar_credito(monto, ingreso, deuda, edad):
  if (edad>25) and (deuda<=monto/2) and (monto<ingreso*3):
    if edad>=65 and deuda>0:
      return False
    else:
      return True
  else:
    return False
```

```python
# cumple con todas las condiciones
print('Estos casos deberian ser True')
print( evaluar_credito(1e6, .4e6, 0, 30) )
print( evaluar_credito(.2e6, .4e6, 0, 70) )
print( evaluar_credito(.3e6, 300000, 0, 40) )

# no cumple con las condiciones
print('Estos casos deberian ser False')
# caso 1: soy mayor de 65 y tengo deuda
print( evaluar_credito(.3e6, 400000, 1000, 70) )

# caso 3: no cumple con edad minima
print( evaluar_credito(.1e6, 400000, 0, 20) )

# caso 4: no cumple por monto maximo
print( evaluar_credito(3e6, 500000, 0, 30) )
```
