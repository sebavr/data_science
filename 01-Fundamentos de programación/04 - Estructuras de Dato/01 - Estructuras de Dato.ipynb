## 4. Set

```python
# colección de elementos únicos
# ordenados en la medida de lo posible
# podemos representar teoria de conjuntos
{3,2,1,2,3,3,2,2,3,1,1,3,2,3}
```

```python
A = {1,2,3}
B = {3,4,5,6}
```

```python
A.union(B)
```

```python
A.intersection(B)
```

```python
A.difference(B)
```

```python
A.isdisjoint(B)
```
