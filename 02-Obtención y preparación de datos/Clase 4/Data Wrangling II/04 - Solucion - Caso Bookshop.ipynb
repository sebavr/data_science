```python
df_web.columns = df_stores.columns
```

```python
df_total = pd.concat([df_web,df_stores])
```

```python
df_total.dropna(inplace=True)
```

```python
df_total
```

```python
df_total.groupby('Title').count()
```

```python
# Regularizando el titulo "Pining For The Fisheries of Yore"
df_total['Title'] = df_total['Title'].apply(lambda x: x.capitalize())
```

```python
df_total.groupby('Title').count()
```

```python
# Regularizando el contenido de la columna Title (el apostrofe)
df_total['Title'] = df_total['Title'].apply(lambda x: x.replace("’","'"))
```

```python
df_total.groupby('Title').count()
```

```python
df_total.groupby('Title').sum()[['Units sold']].sort_values(by='Units sold', ascending=False)
```
