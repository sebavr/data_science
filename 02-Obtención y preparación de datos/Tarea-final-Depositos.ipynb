# Formato a Fecha

```python
def fechaa(f):
    año=f[:4]
    mes=f[5:]
    return mes+'-'+año
```

```python
fechaa('2009m12')
```

```python
fecha_nueva=[]
for i in df['fecha']:
    j=fechaa(i)
    fecha_nueva.append(j)
    print(j)
    
    
```

# Pegar todo en un dataframe

```python
final=pd.DataFrame(columns=['Fecha','Institución Financiera','Tipo','Subtipo','Valor','Tipo de moneda'])
```

```python
def formatNumber(number, decimals, espanol=True):
    if type(number) != int and type(number) != float:
        return number
 
    d={'.':',', ',':'.'}
    return ''.join(d.get(s, s) for s in f"{number:,.{decimals}f}") \
        if espanol \
        else f"{number:,.{decimals}f}"


final['Fecha']=fecha_nueva
final['Institución Financiera']=n_inst
final['Tipo']=n_tipo
final['Subtipo']=n_subtipo
final['Valor']=formatNumber(round(n_valor,0), 0)
final['Tipo de moneda']=n_moneda
```

```python
final
```

```python
df__final=final.pivot_table(values='Valor', index=['Fecha','Institución Financiera'], columns=['Tipo','Subtipo','Tipo de moneda'], aggfunc='sum', fill_value=0)
```
