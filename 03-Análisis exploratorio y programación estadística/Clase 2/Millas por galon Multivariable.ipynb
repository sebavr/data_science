### Seleccion de modelo para las variables

### Modelo con todas las variables

```python
import statsmodels.formula.api as smf
import statsmodels.tools.eval_measures as metrics

lm = smf.ols(formula='''Q("mpg") 
                            ~ Q("displacement") + 
                              Q("horsepower") + 
                              Q("weight") + 
                              Q("acceleration")+
                              Q("model year")+
                              Q("origin")
                              ''',data=df).fit()
```

```python
lm.params
```

```python
lm.summary()
```

```python
y_true = df['mpg']
y_pred = lm.predict(df[['displacement','horsepower','weight','acceleration','model year','origin']])
print( 'MAE: {}'.format(metrics.meanabs(y_true,y_pred)) ) 
print( 'MSE: {}'.format(metrics.mse(y_true,y_pred)) )
print( 'RMSE: {}'.format(metrics.rmse(y_true,y_pred)) )
print( 'R2: {}'.format(lm.rsquared))
print( 'R2-Adj: {}'.format(lm.rsquared_adj))

```

```python
plt.scatter(y_true,y_pred)
plt.xlabel('Y Real')
plt.ylabel('Predicción Y')
```

```python
sns.distplot(y_true-y_pred)
```

- Por cada unidad que se incrementa displacement, el rendimiento promedio se incrementa en 0.012284
- Por cada unidad que se incrementa horsepower, el rendimiento promedio se disminuye en 0.014374
- Por cada unidad que se incrementa weight, el rendimiento promedio se disminuye en 0.006601
- Por cada unidad que se incrementa acceleration, el rendimiento promedio se incrementa en 0.088194
- Por cada unidad que se incrementa model year, el rendimiento promedio se incrementa en 0.751957
- Por cada unidad que se incrementa origin, el rendimiento promedio se incrementa en 1.384268


### Eliminando las variables con alta correlación entre si
### método Backward Elimination

```python
lm = smf.ols(formula='''Q("mpg") 
                            ~ 
                            Q("displacement")+
                            Q("weight")+
                              Q("model year")+
                              Q("origin")
                              ''',data=df).fit()
```

```python
lm.params
```

```python
lm.summary()
```

```python
y_true = df['mpg']
y_pred = lm.predict(df[['displacement','weight','model year','origin']])
print( 'MAE: {}'.format(metrics.meanabs(y_true,y_pred)) ) 
print( 'MSE: {}'.format(metrics.mse(y_true,y_pred)) )
print( 'RMSE: {}'.format(metrics.rmse(y_true,y_pred)) )
print( 'R2: {}'.format(lm.rsquared))
print( 'R2-Adj: {}'.format(lm.rsquared_adj))
```

```python
plt.scatter(y_true,y_pred)
plt.xlabel('Y Real')
plt.ylabel('Predicción Y')
```

```python
sns.distplot(y_true-y_pred)
```

### Eliminando las variables con alta correlación entre si
### método Backward Elimination(2)

```python
lm = smf.ols(formula='''Q("mpg") 
                            ~ 
                            
                            Q("weight")+
                              Q("model year")+
                              Q("origin")
                              ''',data=df).fit()
```

```python
lm.params
```

```python
lm.summary()
```

```python
y_true = df['mpg']
y_pred = lm.predict(df[['weight','model year','origin']])
print( 'MAE: {}'.format(metrics.meanabs(y_true,y_pred)) ) 
print( 'MSE: {}'.format(metrics.mse(y_true,y_pred)) )
print( 'RMSE: {}'.format(metrics.rmse(y_true,y_pred)) )
print( 'R2: {}'.format(lm.rsquared))
print( 'R2-Adj: {}'.format(lm.rsquared_adj))
```

```python
plt.scatter(y_true,y_pred)
plt.xlabel('Y Real')
plt.ylabel('Predicción Y')
```

```python
sns.distplot(y_true-y_pred)
```

```python
El factor que tiene mas incidencia en el modelo es origin, dado que su coeficiente es mayor. Se logra un mayor incremento si este es mas alto.
Es mejor modelo que el simple, ya que se explica una mayor variabilidad de la varianza en el caso multivariado (R2).
```
