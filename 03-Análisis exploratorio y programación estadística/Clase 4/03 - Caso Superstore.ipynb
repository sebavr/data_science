### Limpieza y Wrangling

```python
df.isnull().sum()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df['Order Date'].describe()
```

```python
df['Year']=df['Order Date'].apply(lambda x : x.year)
df['Month']=df['Order Date'].apply(lambda x : x.month)
```

### AnÃ¡lisis por regiÃ³n y segmento

```python
sns.countplot(x='Segment', data=df)
```

```python
sns.countplot(x='Region', data=df)
```

```python
sns.barplot(x='Region', y='Sales', data=df)
```

```python
sns.barplot(x='Segment', y='Sales', data=df)
```

```python
p=df.pivot_table(index='Region',columns='Segment', values='Sales', aggfunc='sum')
p
```

```python
sns.heatmap(p, cmap='coolwarm',linewidths=1,linecolor='white',annot=True, fmt='f')
```

### Ventas por estado y aÃ±o

```python
p2=df.pivot_table(index='State',columns='Year', values='Sales', aggfunc='sum',fill_value=0)
p2
```

```python
fig= plt.figure(figsize=(6,10))
axe=fig.add_axes([0,0,1,1])
sns.heatmap(p2, cmap='coolwarm',linewidths=1,linecolor='white',annot=False, fmt='f',ax=axe)
```

### Ventas por mes

```python
p3=df.pivot_table(index='Month',columns='Year', values='Sales', aggfunc='sum',fill_value=0)
p3
```

```python
fig= plt.figure(figsize=(3,4))
axe=fig.add_axes([0,0,1,1])
sns.heatmap(p3, cmap='coolwarm',linewidths=1,linecolor='white',annot=False, fmt='f',ax=axe)
```

### Analisis de Categorias

```python
p4=df.pivot_table(index='Sub-Category',columns='Category', values='Sales', aggfunc='sum')
p4
```

```python
fig= plt.figure(figsize=(4,4))
axe=fig.add_axes([0,0,1,1])
sns.heatmap(p4, cmap='coolwarm',linewidths=1,linecolor='white',annot=False, fmt='f',ax=axe)
```
