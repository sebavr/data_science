# SVM - Social Network Ads

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('SocialNetworkAds.csv')
```

```python
df.head(2)
```

#### Modelo

```python
# Modelo con Age y Estimated Salary
X = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

#### Escalamiento

```python
from sklearn.preprocessing import StandardScaler
```

```python
# utilizaremos el escalador estandar (estandarización)
scaler = StandardScaler()
scaler.fit(X)
```

```python
X_sc = scaler.transform(X)
```

```python
X_sc[:5]
```

#### Cross-Validation

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sc, y, test_size=0.25, random_state=41)
```

### 1. Clasificador SVM con kernel Lineal

```python
from sklearn.svm import SVC
```

```python
classifier= SVC(kernel='linear', random_state=0)
classifier.fit(X_train,y_train)
```

#### Métricas de Evaluación

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y_pred=classifier.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

#### Visualización Training Set

```python
from matplotlib.colors import ListedColormap

#
# Funcion muy util para graficar los limites de decision
#
def grafica_decision_boudary(X_set, y_set, title):
    # Visualising the Test set results
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                         np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
    plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                 alpha = 0.75, cmap = ListedColormap(('red', 'green')))
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                    c = ListedColormap(('red', 'green'))(i), label = j)
    plt.title(title)
    plt.xlabel('Age')
    plt.ylabel('Estimated Salary')
    plt.legend()
    plt.show()
```

```python
grafica_decision_boudary(X_train,y_train,'SVM linear kernel training set')
```

#### Visualización Test Set

```python
grafica_decision_boudary(X_test,y_test,'SVM linear kernel test set')
```

### 2. Clasificador SVM con kernel RGF

```python
classifier= SVC(kernel='rbf', random_state=0)
classifier.fit(X_train,y_train)
```

#### Métricas de Evaluación

```python
y_pred=classifier.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

#### Visualización Training Set

```python
grafica_decision_boudary(X_train,y_train,'SVM rbd kernel training set')
```

#### Visualización Test Set

```python
grafica_decision_boudary(X_test,y_test,'SVM rbd kernel test set')
```
