# Support Vector Regressor

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('auto-mpg.csv')
```

```python
df.head(2)
```

```python
plt.title('Millas por GalÃ³n de combustible')
sns.scatterplot(data=df, x='weight', y='mpg')
```

```python
plt.title('Millas por GalÃ³n de combustible')
sns.regplot(data=df, x='weight', y='mpg')
```

```python
plt.title('Millas por GalÃ³n de combustible')
sns.regplot(data=df, x='weight', y='mpg', order=2)
```

## Modelo

```python
X = df[['weight']]
y =df['mpg']
```

## Support Cruzada

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)
```

## Support Vector Regressor

```python
from sklearn.svm import SVR
```

```python
regressor = SVR(kernel = 'rbf')
```

```python
regressor.fit(X_train, y_train)
```

## EvaluaciÃ³n

```python
y_pred = regressor.predict(X_test)
```

```python
regressor.score(X_test,y_test)
```

## VisualizaciÃ³n

```python
X_plot = np.arange(1600,5500).reshape(-1,1)
X_plot
```

```python
plt.title('Millas por GalÃ³n de combustible')
plt.plot(X_plot, regressor.predict(X_plot), c='k', label='Modelo Polinomial')
sns.scatterplot(data=df, x='weight', y='mpg', label='Datos Reales')
plt.legend(loc=0)
```
