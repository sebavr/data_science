# Clustering Jerarquico

```python
# imports
from matplotlib import pyplot as plt
import numpy as np
```

```python
# setups
%matplotlib inline
np.set_printoptions(precision=5, suppress=True)  # suprime notación científica en los tipos de dato float
```

### Generación de los datos de ejemplo

```python
from sklearn.datasets import make_blobs
```

```python
# make_blobs retorna una tupla (puntos, centroide al que pertenece)
X, y = make_blobs(n_samples=200, n_features=2, centers=2, cluster_std=1.8, random_state=101)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X[:,0], X[:,1], c=y)
plt.title('Datos de Ejemplo')
```

### Realizando la Clusterización Jerárquica

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#linkage realiza la clusterización
#El método ward utiliza el método de minmizacion de la varianza de Joe Ward
#Otro parametro es la metrica de distancia
Z = linkage(X, method='ward')
```

```python
#Chequeando el coeficiente de correlación Cophenetic
#Es un valor que va desde 0 a 1, mientras más cercano a 1 mejor respeta la distancia
#compara(correlaciona) las distancias reales versus las distancias asignadas en el dendrograma

from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

c, coph_dists = cophenet(Z,pdist(X))
c
```

```python
#Revisemos el primer registro de la matriz de enlaces
#Corresponde a la priemra iteracion del aglomeramiento
#Toma el elemento indice 7 y lo junto con el elemento indice 189, que se encuentran a una distancia de 0.04693
#creando un cluster de 2 elementos
#(idx1, idx2, dist, sample_count)
Z[0]
```

```python
# y el segundo registro
Z[1]
```

```python
#notese que Z[12] referencia el elemento id 208 y solo tenemos 200 elementos
# es decir, ya comienza a enlazar elementos a otros clusters
Z[:20]
```

### Graficando un Dendrograma

```python
#Un dendrograma es una visualizacion en forma de arbol que muestra como se realizó la clusterizacion
# muestra el orden y las distancias
#https://www.mathworks.com/helps/stats/dendrogram.html
plt.figure(figsize=(25,10))
plt.title('Dendrograma de Clustering Jerarquico')
plt.xlabel('Indice de la muestra')
plt.ylabel('Distancia')
dendrogram(
    Z,
    leaf_rotation=90.,
    leaf_font_size=8.,
)
plt.show()
```

```python
Z[-4:]
```

### Personalizaciones al Dendrograma

```python
#Mostrar una cantidad de clusters determinadas
plt.figure(figsize=(25,10))
plt.title('Dendrograma de Clustering Jerarquico')
plt.xlabel('Indice de la muestra')
plt.ylabel('Distancia')
dendrogram(
    Z,
    truncate_mode='lastp',# muestra solo los ultimos p clusters agrupados
    p=5, # cantidad de clusters agrupados a mostrar
    show_leaf_counts=False, # si es True: muestra la cantidad de indices agrupados de cada cluster
    show_contracted=True # para desplegar una distribucion en las ramas truncadas
    
)
plt.show()
```

```python
#Mostrar una cantidad de clusters determinadas
plt.figure(figsize=(25,10))
plt.title('Dendrograma de Clustering Jerarquico')
plt.xlabel('Indice de la muestra')
plt.ylabel('Distancia')
dendrogram(
    Z,
    truncate_mode='lastp',# muestra solo los ultimos p clusters agrupados
    p=10, # cantidad de clusters agrupados a mostrar
    show_leaf_counts=True # si es True: muestra la cantidad de indices agrupados de cada cluster
    #show_contracted=True # para desplegar una distribucion en las ramas truncadas
    
)
plt.show()
```

### Seleccionando la cantidad de clusters

```python
# La mejor forma es fijandose donde esta la mayor distancia entre clusters

max_d=50

#Mostrar una cantidad de clusters determinadas(seleccion manual)
plt.figure()
plt.title('Dendrograma de Clustering Jerarquico')
plt.xlabel('Indice de la muestra')
plt.ylabel('Distancia')
dendrogram(
    Z,
    truncate_mode='lastp',# muestra solo los ultimos p clusters agrupados
    p=10, # cantidad de clusters agrupados a mostrar
    show_leaf_counts=True # si es True: muestra la cantidad de indices agrupados de cada cluster
    #show_contracted=True # para desplegar una distribucion en las ramas truncadas
)
plt.axhline(y=max_d,c='k')
plt.show()
```

### Obteniendo el cluster

```python
from scipy.cluster.hierarchy import fcluster
```

```python
# fijando la distancia de corte
clusters = fcluster(Z, max_d, criterion='distance')
clusters
```

```python
# fijando la cantidad de clusters
clusters = fcluster(Z,2,criterion='maxclust')
clusters
```

### Visualizando el Cluster

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1], c=clusters, cmap='coolwarm')
```
