# Caso Mall Customers - KMeans

Se cuenta con un dataset de clientes de un importante mall de la ciudad. Los datos son los siguientes:

- CustomerID: correlativo identificador del cliente
- Genre: sexo del cliente
- Age: edad del cliente
- Annual Income: sueldo anual en miles de dolares
- Spending Score: score de gasto calculado para el cliente, mientras mayor más gasta

Se busca realizar una caracterización de los clientes, para lo cual se le solicita que realice una **Clusterización Jerárquica** considerando solamente las variables **Annual Income** y **Spending Score**. Se pide lo siguiente:

- Realice una clusterización K-Means
- Aplique método del codo para determinar la cantidad de clusters
- Aplique método de la silueta
- Agregue la clusterización al dataset

---

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Carga de datos

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset.head()
```

### Formulación del Modelo

```python
X = dataset.iloc[:, [3, 4]].values
```

### Escalamiento

```python
from scipy.cluster.vq import whiten
```

```python
scaled_X = whiten(X)
```

```python
scaled_X
```

```python
X=scaled_X
```

```python
plt.figure(figsize=(8,7))
# Completar   ... plt.scatter(...)
p=plt.scatter(X[:,0],X[:,1], c='y', cmap='rainbow');
```

### Clusterización KMeans

```python
from sklearn.cluster import KMeans
```

```python
#Supongamos que partimos con K=5
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
pronostico=kmeans.labels_
pronostico
```

```python
kmeans.inertia_
```

#### Graficar la clusterización

```python
fig, (axes1,axes2) = plt.subplots(1,2, sharey=True, figsize=(18,7))
axes1.set_title('Grupo Estudio')
axes1.scatter(X[:,0],X[:,1],c='y')

axes2.set_title('K Means')
axes2.scatter(X[:,0],X[:,1],c=kmeans.labels_)
axes2.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='r', marker='*', s=200, edgecolors='w')
```

#### Método del Codo

```python
import warnings
warnings.filterwarnings("ignore")
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Metodo del codo')
plt.xlabel('Numero de clusters (k)')
plt.ylabel('Inercia')
# COMPLETAR
```

#### Coeficiente de silueta

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
savg=[]
for i in range(2,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=42)
    kmeans.fit(X)
    silhouette_avg=silhouette_score(X, kmeans.labels_)
    savg.append(silhouette_avg)
```

```python
plt.plot(range(2,11),savg)
plt.title('Metodo del coeficiente de silueta')
plt.xlabel('Numero de clusters (k)')
plt.ylabel('Coef. Silueta')
```

#### Agregando clusterizacion al dataset

```python
def punt(data):
    puntos=data
    if (puntos==4):
        return 'Conservador'
    if (puntos==3):
        return 'Descuidado'
    if (puntos==2):
        return 'Objetivo'
    if (puntos==1):
        return 'Estandar'
    if (puntos==0):
        return 'Cauto'
```

```python
pronostico
```

```python
dataset['puntuacion']=pronostico
```

```python
dataset['caracterizacion']=dataset['puntuacion'].apply(punt)
```
