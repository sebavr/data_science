### Cargar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Lectura de datos

```python
df = pd.read_csv('wine-data.csv')
```

```python
df.head(2)
```

```python
df.shape
```

```python
### Escalamiento de datos
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
```

```python
pd.DataFrame(scaled_data[:5,:],columns=df.columns)
```

```python
### Análisis de Dimensionalidad
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA().fit(scaled_data)
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(range(1,len(pca.explained_variance_ratio_)+1), np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Numero de componentes')
plt.ylabel('Varianza Explicada Acumulada')
plt.title('Varianza Explicada por Componentes Principales')
plt.show()
```

9 componentes explican cerca del 95% de la varianza de los datos

```python
### Reducción de dimensionalidad
```

```python
pca = PCA(n_components=9)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
# Visualizando la matriz de transformacion
componentes = pd.DataFrame(pca.components_,columns=df.columns)
componentes
```

```python
plt.figure(figsize=(18,2))
sns.heatmap(componentes, annot=True, cmap='coolwarm')
```

```python
# Antes de aplicar la reducción de dimensionalidad
scaled_data.shape
```

```python
# Después de aplicar la reducción de dimensionalidad
x_pca.shape
```

```python
c=pd.DataFrame(x_pca,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9'])
c
```

```python
# Visualización de los datos con 9 Componentes
plt.scatter(x_pca[:,0],x_pca[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Visualización de Datos con 9 Componenter Principales')
plt.show()
```

```python
sns.pairplot(c)
```

Se pueden apreciar clusters en ciertas combinaciones como PC1-PC2 donde son mas notorios.    

```python
### Análisis de clusterización
```

```python
from sklearn.cluster import KMeans
```

```python
# Método del Codo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x_pca)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Método del Codo')
plt.xlabel('Número de clusters')
plt.ylabel('Inercia')
plt.show()
```

Serian 4 clusters según metodo del codo

```python
### Clusterización con k=3
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x_pca)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
clases = pd.read_csv('wine-segments.csv')
clases.squeeze()
```

```python
plt.title('Segmentos reales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(x_pca[:,0],x_pca[:,1], c=clases['Cultivar'])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('Clusterización K-Means K=3')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(x_pca[:,0],x_pca[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c='r', marker='*', lw=3)
plt.subplot(1,2,2)
plt.title('Segmentos reales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(x_pca[:,0],x_pca[:,1], c=clases['Cultivar'])
```
