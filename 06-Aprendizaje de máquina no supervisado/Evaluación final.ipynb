```python
### Cargar librerias
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
### Lectura de datos
```

```python
df = pd.read_csv('hespe-data.csv',sep=';')
```

```python
df.head(2)
```

```python
df2=df[['mother_ed','farther_ed','siblings','parental_status','mother_occup','father_occup']]
df2.head(2)
```

```python
df2.shape
```

145 datos y 6 variables

```python
### Escalamiento de datos
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df2)
scaled_data = scaler.transform(df2)
```

```python
pd.DataFrame(scaled_data[:5,:],columns=df2.columns)
```

```python
### Análisis de Dimensionalidad
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA().fit(scaled_data)
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(range(1,len(pca.explained_variance_ratio_)+1), np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Numero de componentes')
plt.ylabel('Varianza Explicada Acumulada')
plt.title('Varianza Explicada por Componentes Principales')
plt.show()
```

- 3 componenentes explican 68% de las varianzas
- 4 componenentes explican 80% de las varianzas
- 5 componenentes explican 92% de las varianzas

```python
### Reducción de dimensionalidad
```

```python
pca = PCA(n_components=4)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
# Visualizando la matriz de transformacion
componentes = pd.DataFrame(pca.components_,columns=df2.columns)
componentes
```

```python
plt.figure(figsize=(18,2))
sns.heatmap(componentes, annot=True, cmap='coolwarm')
```

```python
# Antes de aplicar la reducción de dimensionalidad
scaled_data.shape
```

```python
# Después de aplicar la reducción de dimensionalidad
x_pca.shape
```

```python
c=pd.DataFrame(x_pca,columns=['PC1','PC2','PC3','PC4'])
c
```

```python
sns.pairplot(c)
```

```python
### Análisis de clusterización
```

```python
from sklearn.cluster import KMeans
```

```python
import warnings 
warnings.filterwarnings('ignore')
# Método del Codo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 30)
    kmeans.fit(x_pca)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Método del Codo')
plt.xlabel('Número de clusters')
plt.ylabel('Inercia')
plt.show()
```

7 clusters

```python
### Clusterización con k=4
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(x_pca)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
clases = df[['grade']]
clases.squeeze()
```

```python
plt.title('Segmentos reales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(x_pca[:,0],x_pca[:,1], c=clases['grade'])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('Clusterización K-Means K=6')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(x_pca[:,0],x_pca[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c='r', marker='*', lw=3)
plt.subplot(1,2,2)
plt.title('Segmentos reales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(x_pca[:,0],x_pca[:,1], c=clases['grade'])
```

```python
### Clusterización con Gaussian Mixture
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=5)
gmm.fit(x_pca)
```

```python
y_cluster_gmm = gmm.predict(x_pca)
y_cluster_gmm
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('Clusterización K-Means K=4')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(x_pca[:,2],x_pca[:,1], c=y_cluster_gmm)
plt.subplot(1,2,2)
plt.title('Segmentos reales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(x_pca[:,2],x_pca[:,1], c=clases['grade'])
```
